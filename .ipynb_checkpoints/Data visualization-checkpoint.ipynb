```python
from PIL import Image
import imageio
import os

# Path to image and mask folders
image_folder = 'CVC-ClinicDB/Original'
mask_folder = 'CVC-ClinicDB/Ground Truth'

# Get a list of all image files in the folder
image_files = [f for f in os.listdir(image_folder) if f.endswith('.tif')]

# Iterate through all image files
for image_file in image_files:
    # Construct the full path to the image and mask
    image_path = os.path.join(image_folder, image_file)
    mask_path = os.path.join(mask_folder, image_file)

    # Open the image and mask using imageio
    img_array = imageio.imread(image_path)
    mask_array = imageio.imread(mask_path)

    # Print the size of the image and mask
    print(f"Image Size: {img_array.shape}, Mask Size: {mask_array.shape}")
```
